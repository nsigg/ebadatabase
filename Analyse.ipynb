```python
def analyse_data(data_old, data_new):
    print("Old Data: ")
    print( data_old.columns)
    print("New Data: ")
    print(data_new.columns)
    
    print("Old Data " + str(len(data_old.columns)))
    print("New Data " + str(len(data_new.columns)))
    print("=" * 60)
    
    print("Old Data: ")
    print(data_old.info())
    print("New Data: ")
    print(data_new.info())
    
```

#### Datenvergleich Banks

```python
analyse_data(banks_2020, banks_2021)
```
