## 1.0 Eingabe personalisierter Parameter

```python
DATABASE_NAME = 'ebs'
DATABASE_USER = 'root'
DATABASE_PASSWORD = 'mariadb'

CURRENT_YEAR = 2021
```

## 1.1 Import relevanter Funktionen und Herstellen der Vebindung zur Datenbank

```python
import os
from src.core.CreatDatabaseSchema import setUpDatabase
from src.core.InsertMetadata import InsertMetadata
from src.core.InsertCsvData import InsertCsvData
from src.core.Connection import getConnection, getRootConnection

data_path = os.path.join(os.getcwd(), 'data')
rootConnection = getRootConnection(DATABASE_USER, DATABASE_PASSWORD)
import pandas as pd
import src.core.WebScraping as webScraper
import datetime
import os
```

## 1.2 Setup der Datenbank

```python
rawDataPath = 'data/RawData'    
downloadYear = 2020
preprocessedDataPath = os.path.dirname('data/' + str(downloadYear) + '/')

if not os.path.exists(rawDataPath):
    os.makedirs(rawDataPath)

if not os.path.exists(preprocessedDataPath):
    os.makedirs(preprocessedDataPath)
    webScraper.start_extraction(downloadYear) 
    
print()
print("="*10 +"Scraping Finished"+ "="*10)
print()

for filename in os.listdir(rawDataPath):
    f = os.path.join(rawDataPath, filename)
    
    # checking if it is a file
    if os.path.isfile(f):
        print(filename)
        if filename.__contains__("xlsx"):
            df = pd.read_excel(f,   skiprows=[0])
        else:
            df = pd.read_csv(f, encoding ="latin-1")
        filenameWithoutExtension = filename.split('.')
        csvFilename = preprocessedDataPath + '/' + filenameWithoutExtension[0] + '.csv'
        df.to_csv(csvFilename, encoding='utf-8', index=False)
        
        s = rawDataPath + '/' +filename
        os.remove(s)
print()
print("="*10 +"Preprocessing Finished"+ "="*10)
```

```python
setUpDatabase(rootConnection, DATABASE_NAME)
```

## 1.3 Insert der Daten in die erstellte Datenbank

```python
connection = getConnection(DATABASE_NAME, DATABASE_USER, DATABASE_PASSWORD)

InsertMetadata(connection)
for year in range(2015, CURRENT_YEAR + 1):
    InsertCsvData(year, connection, data_path)
```
